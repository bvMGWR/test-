```python
#identificar las filas precisas con datos nulos 
filas_nulas= datos[datos.isnull().any(axis=1)] #filas 
print(filas_nulas)
```

```python
data_sin_nulos = datos.dropna() #elimina las filas nulas #drop elimina los nulos #este es un dataset paralelo 
data_sin_nulos
```

```python
#funcion para saber si el genero rpg esta presente en la lista }

def filtro_RPG(genres): 
    return "RPG" in genres #nombre tal cual como esta en el dataset y recorre la columna generos 

datosRPG= data_sin_nulos[data_sin_nulos['Genres'].apply(filtro_RPG)]#aqui ocupamos un data set con nulos "datos" deberia usar data_sin_nulos

nombresRPG= datosRPG[["Title", "Rating"]]
nombresRPG.head(10)
```
